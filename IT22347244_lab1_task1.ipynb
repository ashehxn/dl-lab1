```python
from __future__ import division, print_function, unicode_literals
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

exp1 = np.random.exponential(1, (4,4))
print(exp1)

```

```python

exp_data = np.random.exponential(1, (100000, 1))
uni_data = np.random.rand(100000, 1)
nor_data = np.random.randn(100000, 1)

plt.hist(exp_data, density=True, bins=100, histtype="step", color="blue", label="Exponential")
plt.hist(uni_data, density=True, bins=100, histtype="step", color="red", label="Uniform")
plt.hist(nor_data, density=True, bins=100, histtype="step", color="green", label="Normal")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

Z = X**2 + Y**2

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

ax.set_title('3D Plot of Z = X² + Y²')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

plt.show()
```

```python
import seaborn as sns
import pandas as pd

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data = df[features]

pearson_corr = data.corr(method='pearson')
spearman_corr = data.corr(method='spearman')

def plot_heatmap(corr, title):
    plt.figure(figsize=(8, 6))
    sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True,
                cbar_kws={'label': 'Correlation Coefficient'})
    plt.title(title)
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

plot_heatmap(pearson_corr, "Pearson Correlation - Pokémon Stats")
plot_heatmap(spearman_corr, "Spearman Correlation - Pokémon Stats")
```
